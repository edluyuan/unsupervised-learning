```python
# -*- coding: utf-8 -*-

"""
    File name: gibbs_sampler.py
    Description: A re-implementation of the Gibbs sampler for LDA
    Author: Adapted to match MATLAB code structures
    Date created: October 2023
    Python version: 3.6 and above
"""

import numpy as np
import pandas as pd
from scipy.special import gammaln
import matplotlib.pyplot as plt

# Configure matplotlib for better aesthetics
plt.rcParams.update({'font.size': 12})
plt.style.use('seaborn-v0_8-dark')  # Optional: Use a seaborn-like style


class GibbsSampler:
    def __init__(self, D, K, W, alpha, beta, random_seed=None):
        """
        :param D:              Number of documents
        :param K:              Number of topics
        :param W:              Number of words in vocabulary
        :param alpha:          Dirichlet parameter on topic mixing proportions
        :param beta:           Dirichlet parameter on topic word distributions
        :param random_seed:    Random seed for reproducibility
        """
        self.D = D          # Number of documents
        self.K = K          # Number of topics
        self.W = W          # Vocabulary size
        self.alpha = alpha
        self.beta = beta
        self.rand_gen = np.random.RandomState(random_seed)

        # Data structures
        self.I = 0          # Number of (doc, word) pairs
        self.di = []        # Document indices
        self.wi = []        # Word indices
        self.ci = []        # Word counts in training set
        self.citest = []    # Word counts in test set

        self.Id = {}        # Dictionary mapping document to list of indices
        self.Iw = {}        # Dictionary mapping word to list of indices
        self.Nd = np.zeros(self.D, dtype=int)  # Number of words in each document

        # Posterior samples
        self.zi = []        # Topic assignments for each word occurrence

        # Counts
        self.Adk = np.zeros((self.D, self.K), dtype=int)  # Adk(d,k)
        self.Bkw = np.zeros((self.K, self.W), dtype=int)  # Bkw(k,w)

        # Hyperparameters
        self.alpha_vec = np.full(self.K, self.alpha)
        self.beta_vec = np.full(self.W, self.beta)

        # Log-likelihoods
        self.loglike = []
        self.loglike_test = []

    def read_data(self, filename):
        """
        Reads the data file and initializes data structures.
        """
        data = pd.read_csv(filename, sep=' ', header=None, names=['doc', 'word', 'train', 'test'])
        self.I = data.shape[0]
        self.di = data['doc'].values - 1  # Convert to 0-based indexing
        self.wi = data['word'].values - 1
        self.ci = data['train'].values
        self.citest = data['test'].values

        # Initialize Id and Iw
        self.Id = {d: [] for d in range(self.D)}
        self.Iw = {w: [] for w in range(self.W)}
        for i in range(self.I):
            d = self.di[i]
            w = self.wi[i]
            self.Id[d].append(i)
            self.Iw[w].append(i)
            self.Nd[d] += self.ci[i]

    def initialize(self):
        """
        Initializes topic assignments and counts.
        """
        self.zi = []
        for i in range(self.I):
            count = self.ci[i]
            topics = self.rand_gen.randint(0, self.K, size=count)
            self.zi.append(topics)

            # Update counts
            d = self.di[i]
            w = self.wi[i]
            for k in topics:
                self.Adk[d, k] += 1
                self.Bkw[k, w] += 1

    def sample(self, n_iter=100):
        """
        Runs the Gibbs sampler.
        """
        for it in range(n_iter):
            for i in range(self.I):
                d = self.di[i]
                w = self.wi[i]
                count = self.ci[i]
                topics = self.zi[i]

                for n in range(count):
                    k = topics[n]
                    # Decrement counts
                    self.Adk[d, k] -= 1
                    self.Bkw[k, w] -= 1
                    self.Nd[d] -= 1

                    # Compute conditional distribution
                    p_k = (self.Adk[d] + self.alpha_vec) * (self.Bkw[:, w] + self.beta) / \
                          (np.sum(self.Bkw, axis=1) + self.beta * self.W)

                    # Normalize
                    p_k /= np.sum(p_k)

                    # Sample new topic
                    new_k = self.rand_gen.choice(self.K, p=p_k)
                    topics[n] = new_k

                    # Increment counts
                    self.Adk[d, new_k] += 1
                    self.Bkw[new_k, w] += 1
                    self.Nd[d] += 1

            # Compute log-likelihood
            ll = self.compute_log_likelihood()
            self.loglike.append(ll)

            if (it + 1) % 10 == 0 or it == 0:
                print(f"Iteration {it + 1}/{n_iter}, Log-Likelihood: {ll}")

    def compute_log_likelihood(self):
        """
        Computes the log joint probability (log-likelihood) of the data.
        """
        ll = 0
        # Theta part
        for d in range(self.D):
            ll += gammaln(np.sum(self.alpha_vec)) - np.sum(gammaln(self.alpha_vec))
            ll += np.sum(gammaln(self.Adk[d] + self.alpha_vec) - gammaln(self.alpha_vec))
            ll -= gammaln(np.sum(self.Adk[d] + self.alpha_vec)) - gammaln(np.sum(self.alpha_vec))

        # Phi part
        for k in range(self.K):
            ll += gammaln(np.sum(self.beta_vec)) - np.sum(gammaln(self.beta_vec))
            ll += np.sum(gammaln(self.Bkw[k] + self.beta_vec) - gammaln(self.beta_vec))
            ll -= gammaln(np.sum(self.Bkw[k] + self.beta_vec)) - gammaln(np.sum(self.beta_vec))

        return ll

    def get_theta_phi(self):
        """
        Computes theta and phi from the counts.
        """
        theta = (self.Adk + self.alpha_vec)
        theta /= np.sum(theta, axis=1)[:, np.newaxis]

        phi = (self.Bkw + self.beta_vec)
        phi /= np.sum(phi, axis=1)[:, np.newaxis]

        return theta, phi

    def run(self, filename, n_iter=100):
        """
        Runs the Gibbs sampler on the data from the specified file.
        """
        self.read_data(filename)
        self.initialize()
        self.sample(n_iter)
        theta, phi = self.get_theta_phi()
        return theta, phi


def main():
    # Set parameters
    alpha = 1.0
    beta = 1.0
    random_seed = 0

    # Read data to determine D and W
    data = pd.read_csv('toyexample.data', sep=' ', header=None, names=['doc', 'word', 'train', 'test'])
    D = data['doc'].max()
    W = data['word'].max()
    K = 3  # Number of topics

    # Initialize sampler
    sampler = GibbsSampler(D=int(D), K=K, W=int(W), alpha=alpha, beta=beta, random_seed=random_seed)
    # Run sampler
    theta, phi = sampler.run('toyexample.data', n_iter=200)

    # Plot log-likelihood
    plt.figure(figsize=(10, 6))
    plt.plot(sampler.loglike)
    plt.xlabel('Iteration')
    plt.ylabel('Log-Likelihood')
    plt.title('Gibbs Sampler Log-Likelihood')
    plt.show()

    # Display theta and phi
    print("Theta (Document-Topic Distributions):")
    print(theta)
    print("\nPhi (Topic-Word Distributions):")
    print(phi)


if __name__ == '__main__':
    main()
```
